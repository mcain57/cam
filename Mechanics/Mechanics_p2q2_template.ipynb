Import standard set of modules for scientific computing and plotting [ALREADY COMPLETE]:

```python
import numpy as np
import scipy as sp
import math
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
matplotlib.rcParams.update({'font.size': 12,'font.family':'serif'})
```

Initialise specific numbers for this example [COMPLETE THE GAPS]:

```python
# Set up some example parameters to test
m=10
r=5
g=10

# Set up initial conditions
th0    = 0
thdot0 = 1
```

Do the numerical integration [COMPLETE THE GAPS]:

```python
# Need to set up time axis to integrate over
N = np.int16(1e2)
t = np.linspace(0,5,N)
dt = t.max() / N

# initialise variables
th     = np.zeros([N,1])
thdot  = np.zeros([N,1])
thddot = np.zeros([N,1])

# use the initial conditions from above
th[0]    = th0
thdot[0] = thdot0

# carry out the integration [COMPLETE THE GAPS: CHOOSE WHICH SCHEME TO USE]
for k in range(1,N):
    thddot[k-1] = -g/r*np.sin(th[k-1])
    thdot[k]    = thdot[k-1] + thddot[k-1] * dt
    th[k]       = th[k-1] + thdot[k-1] * dt

# compute the solution in degrees and also the reaction force [COMPLETE THE GAPS]
thdeg = th*180/np.pi
reaction_force = m*g*np.cos(th) + m*r*thdot**2
```

Plot [ALREADY COMPLETE]:

```python
plt.figure(figsize=(9.5,6),dpi=100)
plt.grid()
plt.xlim([0,t.max()])
plt.xlabel('$t$ (s)')
plt.ylabel('$\\theta$ (deg)')
plt.title("$\dot{\\theta}_0 = $"+str(thdot0))

plt.plot(t,thdeg,'-',linewidth=2,color = [0, 0, 0.8])
```

```python
plt.figure(figsize=(9.5,6),dpi=100)

plt.grid()
plt.xlim([0,t.max()])
plt.xlabel('$t$ (s)')
plt.ylabel('reaction force (N)')
plt.title("$\dot{\\theta}_0 = $"+str(thdot0))

plt.plot(t,reaction_force,'-',linewidth=2,color = [0, 0, 0.8])
```
